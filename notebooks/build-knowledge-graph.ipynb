# Build Fathom Knowledge Graph

This notebook runs the full 4-tier extraction pipeline to build the Knowledge Graph:

1. **Tier 4 — Metadata Baseline**: Extract author and category triples from CSV
2. **Tiers 1+2 — LLM Extraction**: Extract themes, moods, tropes, settings from descriptions using constrained generation
3. **Tier 3 — Vector Deduplication**: Merge near-duplicate concept nodes
4. **Graph Construction**: Build NetworkX graph and save as GraphML

After running this notebook, the Gradio dashboard will automatically load the KG and enable the Fathom layer.
